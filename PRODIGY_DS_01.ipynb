```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
file_path = '/mnt/data/world_population.csv'
```

```python
df = pd.read_csv('/content/world_population.csv', skiprows=4)
df.head()
```

```python
df_final = df.drop(['Country Code',	'Indicator Name',	'Indicator Code'], axis=1)
```

```python
chosen_country = 'India'

country_data = df_final[df_final['Country Name'] == chosen_country]

year_columns = [col for col in country_data.columns if col.isdigit()]
years = list(map(int, year_columns))
population = country_data[year_columns].values.flatten()

plt.figure(figsize=(10, 5))
plt.bar(years, population, color='red')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title(f'Population of {chosen_country} Over the Years')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(population, bins=20, color='red', edgecolor='white')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.title(f'Histogram of Population Distribution for {chosen_country}')
plt.show()
```
